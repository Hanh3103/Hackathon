```python
df = df.loc[:, (df != 0).any(axis=0)]
```

```python
df = df.loc[:,df.apply(pd.Series.nunique) != 1]
df.describe()
```

```python
#Korrelation matrix
df_drop_first_and_last = df.iloc[:,1:-1]
print(df_drop_first_and_last)
corr = df_drop_first_and_last.corr()
ax = sns.heatmap(corr, annot=True)
#plt.show()
```

```python
## scale the data to the rang between 0 and 1 
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
data_rescaled = scaler.fit_transform(df_drop_first_and_last)
```

```python
#95% of variance
pca = PCA(n_components = 0.95)
pca.fit(data_rescaled)
reduced = pca.transform(data_rescaled)
```

```python
#how to choose number of components
pca = PCA().fit(data_rescaled)
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('number of components')
plt.ylabel('cumulative explained variance')
plt.show()
```
